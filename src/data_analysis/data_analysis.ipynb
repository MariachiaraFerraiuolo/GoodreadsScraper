```python
pivot = df.pivot_table(
    index='writer',       # rows
    values='title',       # what we're counting
    aggfunc='count'       # how we're aggregating
)

# Sort to see most prolific authors first
pivot = pivot.sort_values(by='title', ascending=False)

print(pivot.head(10))
```

```python
pivot['percentage'] = (pivot['title'] / pivot['title'].sum()) * 100
pivot = pivot.sort_values(by='percentage', ascending=False)

print(pivot.head(10))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

# Get top 10 authors
author_counts = df['writer'].value_counts().head(10).reset_index()
author_counts.columns = ['writer', 'count']

# Plot with hue and pastel palette
plt.figure(figsize=(10, 6))
sns.barplot(
    data=author_counts,
    x='count',
    y='writer',
    hue='writer',              # assign hue
    palette='pastel',
    dodge=False,               
    legend=False             
)

# for i, (count, writer) in enumerate(zip(author_counts['count'], author_counts['writer'])):
#     plt.text(count + 0.3, i, str(count), va='center')  # x, y, text

plt.xlabel("Number of Books")
plt.ylabel("Author")
plt.title("Top 10 Most Prolific Authors")
plt.tight_layout()
plt.show()
```
