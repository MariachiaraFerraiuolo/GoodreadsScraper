```python
topic_names['genres'] = topic_names['genres'].str.replace(r'\]$', '', regex=True)
```

```python
grouped = topic_names.groupby(["genres", "year_group"]).sum(numeric_only=True).reset_index()
```

```python
df_melted = grouped.melt(id_vars=["genres", "year_group"], 
                         var_name="topic", 
                         value_name="value")
```

```python
# Sum of topis per genre
genre_topic_strength = topic_names.groupby("genres").sum(numeric_only=True)

# 5 most important topics for each genre
top_topics_by_genre = genre_topic_strength.apply(lambda row: row.sort_values(ascending=False).head(10).index.tolist(), axis=1)

#top topics for the genre: Historical Fiction
print("Top topic per 'Historical Fiction':", top_topics_by_genre["Historical Fiction"])
```

```python


topic_cols = topic_names.columns.difference(["genres", "year_group"])

#Calculating the 5 top topics per genre
top_topics_by_genre = topic_names.groupby("genres")[topic_cols].sum().apply(
    lambda row: row.sort_values(ascending=False).head(5).index.tolist(), axis=1
)

for genre in topic_names["genres"].unique():
    top_topics = top_topics_by_genre[genre]
    
    filtered_df = topic_names[topic_names["genres"] == genre][["year_group"] + top_topics].fillna(0)
    df_melted = filtered_df.melt(id_vars="year_group", value_vars=top_topics,
                                 var_name="topic", value_name="value")
    
    # Plot
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df_melted, x="year_group", y="value", hue="topic", marker="o")
    plt.title(f"Evolution of the 5 top topics per genre: {genre}", fontsize=16)
    plt.xlabel("Time")
    plt.ylabel("Topic intensity")
    plt.xticks(rotation=45)
    plt.legend(title="Topic", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show() 

    time.sleep(5)
```

```python

# 1. Counting the most frequent genres
top_genres = topic_names['genres'].value_counts().head(5).index.tolist()

# 2. Find the main topic for each of them
topic_cols = topic_names.columns.difference(["genres", "year_group"])

main_topic_by_genre = {}

for genre in top_genres:
    genre_data = topic_names[topic_names['genres'] == genre]
    topic_sums = genre_data[topic_cols].sum()
    main_topic = topic_sums.idxmax()  
    main_topic_by_genre[genre] = main_topic

# 3. Data preparation
plot_data = pd.DataFrame()

for genre, topic in main_topic_by_genre.items():
    genre_df = topic_names[topic_names['genres'] == genre][['year_group', topic]].copy()
    genre_df = genre_df.groupby('year_group')[topic].sum().reset_index()
    genre_df.rename(columns={topic: 'value'}, inplace=True)
    genre_df['genre'] = genre
    genre_df['topic'] = topic
    plot_data = pd.concat([plot_data, genre_df])
# 4. Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=plot_data, x='year_group', y='value', hue='genre', marker='o')


# Aggiunte grafiche
plt.title("Evolution of the main topic for the five most loved genre", fontsize=16)
plt.xlabel("Time")
plt.ylabel("Intensity of the main topic")
plt.xticks(rotation=45)
plt.legend(title="Genere", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
